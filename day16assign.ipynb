```python
df['gender'].value_counts()
```

```python
label = list(df['race/ethnicity'].value_counts().keys())
label
```

```python
value = list(df['race/ethnicity'].value_counts().values)
value
```

```python
plt.bar(label, value)
plt.xticks(rotation = 0)
plt.title('GRAPH')
plt.ylabel('score')
plt.xlabel('race/ethnicity')      #rotates the labels mentioned under
plt.show()
```

```python
plt.pie(value, labels=label, wedgeprops= {'edgecolor':'k'})    #k is short for black
plt.show()
```

```python
list1 = [0.1,0.1,0.1,0.2,0.1]
sns.set(rc={'figure.figsize':(4,4)})
plt.pie(value, labels=label, wedgeprops= {'edgecolor':'k'}, explode=list1, autopct='%1.2f%%', shadow=True)
plt.title('pie chart')
plt.show()
```
